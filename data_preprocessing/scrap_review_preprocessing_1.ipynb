```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
pd.set_option('display.max_colwidth', -1)
```

```python
# scrap 한 파일 불러오기
df = pd.read_csv("C:\Users\user\Desktop\NLP_project\데이터 정리\movie_review_2017_2020.csv", unicode = UCF8)
```

```python
# 데이터 전처리
df = df.reset_index() 
df.dropna(inplace = True)  # nan값 제거
df.drop_duplicates(inplace=True) # 중복값제거
df = df[df.title != 'title']  # title이라는 제목의 데이터 삭제
# title이 특수문자인 경우 삭제
df['title'] = df['title'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣a-zA-Z0-9_]","") 
df.reset_index(inplace=True)
df.drop('index', axis= 1, inplace=True)
# 고정 단어 '추천하였습니다.' review 삭제
drop_index = df[df['review'] == '추천하였습니다.'].index
df = df.drop(drop_index)
```

```python
# 전처리한 데이터 저장
df.to_csv('movie_review_16.csv')
```
